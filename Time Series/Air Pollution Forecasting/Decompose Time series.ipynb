```python
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
filepath = "F:\\All_Prep\\Deep_learning\\Time Series\\Air Pollution Forecasting\\international-airline-passengers.csv"
series = Series.from_csv(filepath, header=0)
result = seasonal_decompose(series, model='additive') #model="multiplicative"
# result
# result.plot()
result.observed.plot(figsize=(20,5),title="observed")
pyplot.show()
result.trend.plot(figsize=(20,5),title="trend")
pyplot.show()
result.seasonal.plot(figsize=(20,5),title="seasonality")
pyplot.show()
result.resid.plot(figsize=(20,5),title="Noice")
pyplot.show()
```
